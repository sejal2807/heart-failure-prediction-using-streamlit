```python
#Interquartile Range(IGQ)
#IQR = q3(75) - q1(25)
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
```

```python
IQR = q3-q1
```

```python
((df < (q1 - 1.5 *IQR)) | (df > (q3 + 1.5 * IQR)))
```

```python
outliers_indices = []

for i in range(df.shape[0]):
        if(~((df < (q1 - 1.5 *IQR)) | (df > (q3 + 1.5 * IQR))).loc[i].any()):
            outliers_indices.append(i)
```

```python
outliers_indices
```

```python
len(outliers_indices)
```

```python
df1 = df.loc[outliers_indices]
```

```python
df1.shape
```

```python
plt.figure(figsize=(16,14))

for i in range(len(df.columns)):
    plt.subplot(4,4,i+1)
    sns.boxplot(df1[df1.columns[i]])
```

```python
#model using logistic , KNN , desiciontree
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from  sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(xfeatures,ylabels,test_size=0.30,random_state=7)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_model.predict(x_test)
```

```python
y_test.values
```

```python
lr_model.score(x_test,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_test,y_test)
```
